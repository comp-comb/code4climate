# Some advanced formatting in `print()` statements

You have written simple `print()` statements. Now we will talk about some advanced ways to format the output of the `print()` statements in Python. We will be using something called `f-string`, which is a feature of newer versions of Python (version 3.6 and above; Google colab uses python 3.11).

The `f-string` allows you to format selected parts of a string.

To specify a string as an `f-string`, simply put an f in front of the string, like this:

```python
#A simple string
original_txt="The price is 49 dollars"
print(original_txt)

#An f-string containing same text
fstring_txt = f"The price is 49 dollars"
print(fstring_txt)
```

### Placeholders and Modifiers

To format values in an `f-string`, add placeholders `{}`. A placeholder can contain variables, operations, functions, and modifiers to format the value.

For example, to print price of an item, instead of using the following
```python
price = 59
print("The price is", price, "dollars")
```
We can use the variable `price` directly into the line to be printed by using `fstring` and `{}`

```python
price = 59
price_txt = f"The price is {price} dollars"
print(price_txt)
# or alteratively,
print(f"The price is {price} dollars")
```
Note how `{price}` in the second example is going to fetch the value in the variable `price`.

# How to get out of a loop: Break and Continue



### Break
The `break` keyword is used to break out a `for` or `while` loop.

For example, let's say we want to end a loop if the square of the iterator `i` is larger than 12, we can do:
```python
for i in range(10):
  if i*i > 12:
    break
  print(i)
```

### Revisiting Secret Number game

We can use the break keyword to improve our secret number game.


```python
#Example code

# What's my secret number?
# Set the secret number and maximum attempts
secret_number = 18
max_attempts = 5

# Start the guessing game
for current_attempt in range(1, max_attempts + 1):
    guess = int(input("Guess the secret number (1-20)"))
    if guess == secret_number:
        print("Congratulations! You guessed the secret number!")
        break
    elif guess > secret_number:
        print("Your guess is too high. Try again.")
        print(f"This was your attempt number: {current_attempt}")
    else:
        print("Your guess is too low. Try again.")
        print(f"This was your  attempt number: {current_attempt}")
```

### Continue

The `continue` keyword skips the iteration, but continue with the next iteration.
For example, if we want to skip printing when the number is 3 but print for all other cases, we  can do

```python
for i in range(10):
  if i == 3:
    continue
  print(i)
```

# Modules (aka Libraries or Packages) in Python



In Python, modules (also called packages or libraries) contain functions and other useful blocks of code that can help extend our code and allow us to do more complex things.

Some of these functions we could write ourselves, but modules are nice in that they are *reusable*. We can add them to any Python code and use them anywhere!

There are many different types of modules, but these are some of the most popular ones:

- `time`: This module provides various time-related functions. It is useful for getting the current time, measuring the time taken by a program to execute, and more.

- `math`: This module provides access to the mathematical functions defined by the `C` standard.

- `matplotlib`: This module is used for data visualization. Almost all plotting and vizualization are done using `matplotlib`.

- `numpy`: This module is used for numerical computing with Python. It provides an array object which can be thought of as a more versatile form of a python list. It also provides some unique tool to reorganize and restructure any list or array object. In simple term, numpy array gives us more flexibility to work with lists.

- `pandas`: This module is used for data manipulation and analysis. It provides data structures for efficiently storing and manipulating large datasets.

We are going to begin our discussion with **`matplotlib`** and then we will discuss **`numpy`**. We will cover **`pandas`** tomorrow.

### Arrays
A numpy array is a grid of values, all of the *same type*, and is indexed by  nonnegative integers starting with zero. The number of dimensions is called the *rank* of the array; the shape of an array is a set of integers giving the size of the array along each dimension.



**NOTE**: Both python lists and numpy array resembles same structures, but the idea of dimensionality in numpy arrays gives the user more control. However, we will not be delving too deep into this.


We can initialize numpy arrays from *nested* (i.e., one within another) Python lists, and access elements using square brackets:


````python
# Numpy array example:

numpy_array_A = np.array([1, 2, 3])
print("Numpy array:", numpy_array_A)

list_A = [1, 2, 3]
print("Python list:",list_A)
````

### Numpy and Python list
Numpy array and Python list are similar but numpy provides mathematical advantage.

````python
# A python list can contain different data types, even another numpy array:

list_B = [1, "Hello", np.array([1, 2, 3])]
print(list_B)

# # But a numpy array cannot store different data types:

# numpy_array_B = np.array([1, "Hello", np.array([1, 2, 3])])
# print(numpy_array_B)
````

````python
# We can convert python list to numpy array:

list_C = [1, 2, 3, "Hello"]
numpy_array_C = np.array(list_C)
print("original list: ", list_C)

print("converted numpy array: ", numpy_array_C)

# Notice any difference???
````

````python
# some basic operations with numpy array
import numpy as np        # we can avoid typing numpy everytime by creating a shorthand for numpy, i.e. np

a = np.array([1, 2, 3])   # Create an array
print(type(a))            # Prints "<class 'numpy.ndarray'>"
print(a.shape)            # Prints "(3,)"
print(a[0], a[1], a[2])   # Prints "1 2 3"
a[0] = 5                  # Change an element of the array
print(a)                  # Prints "[5, 2, 3]"
````


###Mathematical Operations

You can perform mathematical operations on arrays using NumPy functions. For example, `a = np.array([1, 2, 3])` and `b = np.array([4, 5, 6])`, then `c = a + b` creates a new array with the values `[5, 7, 9]`

````python
# we can perform mathametical operation on numpy arrays:

sample_array=np.array([1,2,3])
# sample_array=[1,2,3]

multiplication_array = sample_array * 5
print("Multiplied by 5: ", multiplication_array)
````

````python
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = a + 2*b

print(c)
````

````python
import numpy as np

x = np.array([1,2])
y = np.array([5,6])

# Elementwise sum; both produce the array
print(x + y)
print(np.add(x, y))

# Elementwise difference; both produce the array
print(x - y)
print(np.subtract(x, y))

# Elementwise product; both produce the array
print(x * y)
print(np.multiply(x, y))

# Elementwise division; both produce the array
print(x / y)
print(np.divide(x, y))

# Elementwise square root; produces the array
print(np.sqrt(x))
````

Some common funcitons of numpy that is widely used in data analysis
```

median(): Calculates the median of an array.

std(): Calculates the standard deviation of an array.

var(): Calculates the variance of an array.

min(): Returns the minimum value of an array.

max(): Returns the maximum value of an array.

argmin(): Returns the index of the minimum value in an array.

argmax(): Returns the index of the maximum value in an array.
```



````python
import numpy as np

a = np.array([1, 2, 3, 4, 5])

print(np.mean(a))
print(np.median(a))
print(np.std(a))
print(np.min(a))
print(np.max(a))
print(np.argmin(a))
print(np.argmax(a))
````
