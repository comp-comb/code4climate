However, even though the air quality index is reported as a **single number** for an entire area (ex. Milwaukee is given one AQI value) the effect of AQI on people can depend on many factors including:

and other factors such as exact location and many more!

### Wednesday: Visualizing Air Quality Exposure Risk Throughout the Day

Concept Focus: Lists, Loops, Conditional Statements, plotting with matplotlib

For our project, we will be focusing on the effect the *time of day* can have on a person's exposure to the effects of air quality.

We will be using a sample estimate using the following equation:

`Exposure Index = AQI*RF`

Where RF is a risk factor (from 0 to 1) which can be calculated as follows:

| Time of Day | Risk Factor, RF |
|-------------|-----------------|
|From after 12am to 3am | 0.05|
|From after 3am to 7am| 0.3|
|From after 7am to 10am|0.7|
|From after 10am to 2pm|0.6|
|From after 2pm to 6pm|0.83|
|From after 6pm to 10pm|0.4|
|From after 10pm to 12am|0.22|

````python
import matplotlib.pyplot as plt

AQI = [45, 55, 75, 95, 105, 115, 135, 155, 175, 205, 255, 305, 350, 400, 450, 500, 45, 55, 75, 95, 105, 115, 135, 155, 175]
Time = range(0,24)
EI = list()

for i in Time:

  if i > 0 and i <= 3:

    RF = 0.05

  elif i > 3 and i <= 7:

    RF = 0.3

  elif i > 7 and i <= 10:

    RF = 0.7

  elif i >10 and i <= 2:

    RF = 0.6

  elif i > 2 and i <= 6:

    RF = 0.83

  elif i > 6 and i <=10:

    RF = 0.4

  elif i > 10 or i <= 0:

    RF = 0.22

  EI.append(AQI[i]*RF)

plt.plot(Time, EI, "o--", color="blue")
plt.grid()
plt.title("Exposure Index vs Hours since Midnight")
plt.xlabel("Hours since Midnight")
plt.ylabel("Exposure Index (EI)")

````
