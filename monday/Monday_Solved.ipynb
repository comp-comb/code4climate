## Comments in Our Code

We put comments in our code so that it makes it easier *for us* to understand. Computers do not read or interpret comments, they merely skip those portions of the program. Comments are useful for us to understand the code when we look at it at a later time.

`#` is used for *inline comments*

An *inline comment* means the computer will read that line *up to* the comment character (#) and anything after the comment on that line will not be included.

You can use *inline comments* to comment out an entire line of code (by putting `#` at the begging of a line) or write a comment immediately after some instructions on a line of code.






````python
# This entire line is commented
# This line is commented too

print("anything after # in a single line is commented or ignored by the python program") # Only this portion is commented too
````

# Variables and Data

## Variables

Variables store information. For example, let's look at the statement that you used earlier

```
name = input("Enter your name: ")
```
Here *name* in `name = ` is a variable that will store the name you type in. Variables can be of different types (e.g., letter or number or something else).

Variables in Python are defined and *typed for you* (i.e. Python automatically knows what type of variable it is) when you set or assign a value to them.

In Python, variable names must follow certain rules:


*   A variable name must start with a letter or the underscore (_) character.
*   A variable name *cannot* start with a number.
*   A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _).
*   Variable names are case-sensitive (age, Age and AGE are *three different variables*!)
* You cannot name variables the same thing as certain key words used by python (examples: for, and, or, return, while, etc)

---

***Note:*** Always choose variable names so that, the name itself makes sense with what they are going to be used for (e.g., `name` in the example). Do not choose any variable names randomly (e.g., `abc1`) so that at a later time when you are looking back at the code, it is not intuitive where and why that particular variable is used.


***Note:*** Also, avoid mixing uppercase and lowercase characters when declaring a variable. We recommend using **lowercase** characaters. **Underscores** can also be used to break up long variable names.

---



*Good variable name selection. Human readable and consistent formating*
```
noun = 'chair'
NUMBER_1 = 24
year = 2020
```

*bad variable name selection*
```
NoUN = 'chair'
Number1 = 24
sdfas = 2020
```

## Escape characters:
Some special characters cannot be written directly because they are already used for a different purose by Python. For example, if you want to actually store the character `'`, and you write `c='''` Python will get confused!
For example, try assigining the following statement to a string:
`Tom's response was, "I don't know"`

````python
a_string="Tom's response was, "I don't know""
print(a_string)
````

This is because Python uses `'` or `"` as the start of or at the end of a definition of a string. If you have a `'` or `"` inside a string, it will get confused. This is where **escape characters** come into play. The escape character is denoted by a `backslash (\)`, if you put this `\` in front of `'` or `"`, Python will treat it as a quotation mark instead of the beginning or end of a string.

Try now:

````python
a_string="Tom\'s response was, \"I don't know"
print(a_string)
````

Other common escape characters are
 * `\n` : indicates a new line (aka `linebreak` or `newline`)
 * `\\` : the backslash character itself
 * `\t' : indicates a tab

### Practice Problem 5

Pollutant concentrations are key in calculating air quality and other climate related quantities. These concentrations are often measured in units such as micrograms per cubic meter ($\mu$g/m$^3$) or milligrams per cubic meter (mg/m$^3$).

Write a program that asks the user for a concentration of a pollutant in micrograms per cubic meter and converts that to milligrams per cubic meter (mg/m$^3$) and print the result.

*Note*: 1 gram = 1000000 micrograms and 1 gram = 1000 miligrams


````python
# Convert from microgram per cubic meter to milligram per cubic meter
# Ask the user for the concentration in micrograms per cubic meter
micrograms = float(input("Enter the concentration in micrograms per cubic meter: "))

# Convert to milligrams per cubic meter
milligrams = micrograms / 1000

# Print the result
print("The concentration in milligrams per cubic meter is:", milligrams)
````

### Practice Problem 6

Write a program that reads the temperature in Fahrenheit from a user and converts it to Celsius using the formula below.

$$\frac{C}{5} = \frac{F-32}{9}$$

````python
# F to C conversion
# Convert Fahrenheit to Celsius
fahrenheit = float(input("Enter the temperature in Fahrenheit: "))

# Convert to Celsius
celsius = (fahrenheit - 32) * 5 / 9

# Print the result
print("The temperature in Celsius is:", celsius)
````
