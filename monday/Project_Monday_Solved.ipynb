# Connecting Back to Climate and Air: Calculating Heat Index

Write a program to calculate heat index from temperature (`T`) and relative humidty (`RH`) using the simplified formula.

**Note:** This is an *approximate* formula that is only valid for temperature **between 80 F and 110 F and relative humidity of 40% and above**. See the link below for full range of validity.

$Heat\_Index = -42.379 + (2.04901523 \times T) + (10.14333127 \times RH) - (0.22475541 \times T \times RH) - (0.00683783 \times T^2) - (0.05481717 \times RH^2) + (0.00122874 \times T^2 \times RH) + (0.00085282 \times T \times RH^2) - (0.00000199 \times T^2 \times RH^2)$

Your program should ask the user to enter the temperature (in Fahrenheit) and relative humidity.

Formula from: https://www.weather.gov/ama/heatindex and more details can be found here: https://www.wpc.ncep.noaa.gov/html/heatindex_equation.shtml.


```python
# Ask user for temperature and relative humidity
temperature = float(input("Enter the temperature in Fahrenheit (between 80°F and 110°F): "))
humidity = float(input("Enter the relative humidity (40% and above): "))

# Calculate heat index
heat_index = (-42.379 + (2.04901523 * temperature) + (10.14333127 * humidity)
              - (0.22475541 * temperature * humidity) - (0.00683783 * temperature**2)
              - (0.05481717 * humidity**2) + (0.00122874 * temperature**2 * humidity)
              + (0.00085282 * temperature * humidity**2) - (0.00000199 * temperature**2 * humidity**2))

# Display the result
print(f"The calculated heat index is: {heat_index:.2f}")
```
