# Python Formatting: Whitespace matters!

The code structures we will be running today are run in *blocks* and need to have a way to be seperated from each other. As Python reads through the code you have written, it needs to understand when one set of instruction starts and when it ends (remember that you need to have matching open and closing parenthesis or open and closing quotes on your `print` and `input` statements yesterday). If the set of instruction spans over multiple lines, then we need a way to separate one block of instruction from another. Python does so by `indent`ing the code (some languages use `{ }` for the same purpose). All lines with the same indentation are part of the same block/scope. Therefore, it is absolutely important to avoid any unintentional whitespaces while writing code in python.

```
First set of instruction
    starts the first step
    do the second step
    finish first set of instruction

Second set of instruction
    starts the first step of second set
    do the second step of second set
    finish second set of instruction  
```

The advantage of such approach is that, you need to type less and the code is automatically **made beautifully arranged** by the structure itself.

To add whitespace you can either use **four spaces** or the **tab** key, it does not matter which one you choose as long as you are consistent!

## Mini Project: Calculate Electricity Bill

Calculate the electric bill of a household. The program should ask for the power consumption in KWh (unit of power consumption). Then it will determine the rates as per the table below (the cost of electricity increases incrementally as the usage increases). Finally, it will calculate the total electricity cost.



```
Cost[$/kWh]              Usage [kWh]
    1                        <50
   1.5                  <=50 and <80
    2                   <=80 and <200
   2.5                    above 200
```




````python
# Block for electricity bill
# Calculate electricity bill based on power consumption
power_consumption = float(input("Enter the power consumption in kWh: "))

if power_consumption < 50:
    cost = power_consumption * 1
elif power_consumption <= 80:
    cost = 50 * 1 + (power_consumption - 50) * 1.5
elif power_consumption <= 200:
    cost = 50 * 1 + 30 * 1.5 + (power_consumption - 80) * 2
else:
    cost = 50 * 1 + 30 * 1.5 + 120 * 2 + (power_consumption - 200) * 2.5

print("The total electricity bill is: $", cost)
````

# Loops or Iterators
A loop is used to for when we have some instructions need to be used repeatedly or a series of stuff. Loops help **automate** our code. There are two main types of loops:
  * `for` loop
  * `while` loop

Let's start with the `for` loop
```python
for i in range(1,6):
     print(i)
```
In this, the variable `i` is called by many names: loop counter, or iterator variable, or loop variable, etc.

`range()` : the range function gives us values *up to, but not including* within which the loop counter (`i`) will loop.

We can also loop over the elements of a list or string.

```python
for i in [1,2,3,4,5]:
  print(i)
```
How is the above example different from using the range(1,5) function?

Try both of them and see for yourself.



````python
# Try it out yourself
for i in range(1,6):
     print(i)
````

````python
list_a = [1,2,3,4,5]

for i in list_a:        # the variable 'i' takes one element in each iteration of the loop
  print(i)              # and this variable can be used in algorithmic calculations
````

`for()` and `while()` are very similar, but depending on the situation one may be more convenient than other

The same example with the `while` loop will look like

```python
i = 1
while i < 6:
    print(i)
    i = i+1
```
Note that unlike the `for` loop, the `while` loop construct requires an increment of the counter variable and logical condition for stopping criteria. The loop runs continuously until a given logical condition is met.

Try it yourself below.

**Note** The last statement `i=i+1` is often written in shorthand as `i+=1`

````python
# While loop example:
i = 1
while i < 6:
    print(i)
    i = i+1
````

# Lets play a game: What's your secret number?

Let's devise a game where you will secretly select a lucky number and ask users to guess the number within a few attempts.
Here are the steps:
* First, choose a lucky number (an integer between 1 and 20).
* Store it in a variable in your code
* Fix the number of attempts you will allow an user to try to guess your lucky number.
* Start a loop
  * Ask for a guess from the user
    * If the guess matches with your lucky number, great! Print congratulatory message and game ends.
    * If the guess is not correct one,
      * If the guess is larger than your lucky number, tell user that they guessed a larger number
      * If the guess is smaller than your lucky number, tell user that they guessed a smaller number.
      * Ask the user to try again.
* Continue the loop until the number of attempts hits the limit set by you in the code. In this case, print a message to indicate that the user has reached the limit of attempts.

````python

# Set the secret number and maximum attempts
secret_number = 18
max_attempts = 5

# Start the guessing game
for current_attempt in range(1, max_attempts + 1):
    guess = int(input("Guess the secret number (1-20)"))

    if guess == secret_number:
        print("Congratulations! You guessed the secret number!")
        break
    elif guess > secret_number:
        print("Your guess is too high. Try again.")
    else:
        print("Your guess is too low. Try again.")
else:
    print("Sorry, you've used all your attempts. Better luck next time!")
````

# Functions

A Python function is a block of code that performs a specific task and can be reused in a program. There are two types of functions in Python:
* built-in functions that are provided by Python itself. For example, you already used `print()` or `input()`, and
* user-defined functions that are created by the programmer. This is the one we will be discussing now.

To define a function, you need to use the `def` keyword, followed by a name of your choosing (function-name) and parentheses and colon. For example:

```python
# This is the definition of the function with name greet
def greet():
  print("Hello from a function")
  print("This completes the function greet()")
```

* **Note the colon** `:` after the parenthesis, and
* **the whitespace (or indentation)** after the definition that defines the content of the function.


And when you *call* that function, you do it this way

```python
# And this is where we call the function
greet()
```

For a simple example:

```python
# Let's define a function called Division that takes two numbers and divides the first one by the second and prints out the result
def Division(a, b):
    print(a/b)

# Now we can use this function as many times as we want
Division(3,4)
Division(3.0,4.0)
Division(3,4.0)
Division(3.0,4)
```

Let's try it out

````python
# First function demo
# This is the definition of the function with name greet
def greet():
  print("Hello from a function")
  print("This completes the function greet()")

#Call the function greet
greet()
````

A few more things:

* Functions can be written in the same python script as all of your other code, but functions must be written **before** you call them in your code!
* Functions can have multiple inputs, a single input, or no input as shown above.
* Functions can have outputs using the `return` keyword, where the output is *returned* to a variable or they can have no outputs
* You can also use functions inside of functions. For example, you can print() something out inside of a function
* Inside a function, Python only knows about the variables you define inside the function or pass to it as inputs, it will not be able to use any other variables.


```python
def Division(a, b):
    return(a/b)

# Notice how the whitespaces are now removed to indicate end of the function
# Now we will call the function
result = Division (4,2) # the output of Division will be returned to the variable 'result'
print(result)
```

### Practice Problem 9

**The FizzBuzz Game**

Write a program that prints the numbers from 1 to 100. But for multiples of three, print “Fizz” instead of the number; and for the multiples of five, print “Buzz”. For numbers which are multiples of both three and five print “FizzBuzz”.

Here is some psuedocode (instructional steps) to help get you started:
1. We first want to loop through all of the numbers
2. Within our loop, we will set up a conditional statement (if, elif, else) to check if the number is a multiple of 3,5, or 3 and 5 (what order should we check these conditions in our loop? Does it matter?)
3. Add your print statements inside the if statement


````python
# Fizz Buzz Game
for num in range(1, 101):
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)
````
