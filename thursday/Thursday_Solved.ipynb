### Arrays
A numpy array is a grid of values, all of the *same type*, and is indexed by  nonnegative integers starting with zero. The number of dimensions is called the *rank* of the array; the shape of an array is a set of integers giving the size of the array along each dimension.



**NOTE**: Both python lists and numpy array resembles same structures, but the idea of dimensionality in numpy arrays gives the user more control. However, we will not be delving too deep into this.


We can initialize numpy arrays from *nested* (i.e., one within another) Python lists, and access elements using square brackets:


````python
# Numpy array example:
import numpy as np  # importing numpy again to ensure it's available
numpy_array_A = np.array([1, 2, 3])
print("Numpy array:", numpy_array_A)

list_A = [1, 2, 3]
print("Python list:",list_A)
````


###Mathematical Operations

You can perform mathematical operations on arrays using NumPy functions. For example, `a = np.array([1, 2, 3])` and `b = np.array([4, 5, 6])`, then `c = a + b` creates a new array with the values `[5, 7, 9]`

````python
# we can perform mathametical operation on numpy arrays:

sample_array=np.array([1,2,3])
# sample_array=[1,2,3]

multiplication_array = sample_array * 5
print("Multiplied by 5: ", multiplication_array)
````

````python
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = a + 2*b

print(c)
````

````python
import numpy as np

x = np.array([1,2])
y = np.array([5,6])

# Elementwise sum; both produce the array
print(x + y)
print(np.add(x, y))

# Elementwise difference; both produce the array
print(x - y)
print(np.subtract(x, y))

# Elementwise product; both produce the array
print(x * y)
print(np.multiply(x, y))

# Elementwise division; both produce the array
print(x / y)
print(np.divide(x, y))

# Elementwise square root; produces the array
print(np.sqrt(x))
````

Some common funcitons of numpy that is widely used in data analysis
```
mean(): Calculates the mean of an array.

median(): Calculates the median of an array.

std(): Calculates the standard deviation of an array.

var(): Calculates the variance of an array.

min(): Returns the minimum value of an array.

max(): Returns the maximum value of an array.

argmin(): Returns the index of the minimum value in an array.

argmax(): Returns the index of the maximum value in an array.
```



````python
import numpy as np

a = np.array([1, 2, 3, 4, 5])

print(np.mean(a))
print(np.median(a))
print(np.std(a))
print(np.min(a))
print(np.max(a))
print(np.argmin(a))
print(np.argmax(a))
````

# Pandas: Working with Tables (DataFrames)

Imagine you have a big table of information, like a spreadsheet. `pandas` helps us work with these tables in Python! In `pandas`, these tables are called **DataFrames**. They're super useful for organizing and analyzing data, like the air quality data you'll be using.

Of course, we will have to import `pandas` first. Let's say we will use `pd` as the nickname for `pandas` in our code. So,

````python
import pandas as pd
````

### Creating a DataFrame

You can create a DataFrame using the `pd.DataFrame()` function. One way is to use a Python *dictionary*, where the keys become the column names and the values are lists of the data in each column. We have not talked about *dictionary* before, just know that a *dictionary* is a way to store paired data often referred to as key-value pair written as `key:value`. We will not discuss more about dictionary in this course.

For example:

````python
import pandas as pd

# Create a dictionary with some data
data = {'Name': ['John', 'Mike', 'Sara'],
        'Age': [25, 30, 28],
        'City': ['New York', 'London', 'Paris']}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

# Display the DataFrame
display(df)
````

In the above DataFrame, 0, 1, 2, etc are called the index.

### Selecting a Single Column

To select a single column from a DataFrame, you can use square brackets `[]` with the column name inside. This will give you a pandas Series, which is like a single column of a DataFrame.

````python
# Select the 'Name' column
names = df['Name']

# Display the selected column
display(names)
````

### Selecting Multiple Columns

To select more than one column from a DataFrame, you pass a *list* of the column names you want inside two square brackets `[[]]`. This will give you back a DataFrame containing only those columns.

````python
# Select the 'Name' and 'Age' columns
name_and_age = df[['Name', 'Age']]

# Display the selected columns
display(name_and_age)
````

### Converting a Column to a Python List

Sometimes you might need to work with the data in a column as a regular Python list instead of a pandas Series. You can easily convert a column to a list using the `.tolist()` method. You can also convert it to numpy array using np.array().

````python
# Convert the 'Name' column (which is a Series) to a Python list or numpy array
names_list = names.tolist()
age_array = np.array(df['Age'])



# Display the Python list
print(names_list)
print(age_array)
````

### Saving DataFrames to CSV

Saving your DataFrame to a CSV (Comma Separated Values) file is a common and useful task. CSV files are plain text files that can be easily read by many other programs, including spreadsheet software like Excel or Google Sheets, and other data analysis tools. This allows you to share your data or continue working with it outside of your Python environment.

The `.to_csv()` method is used for this purpose. The `index=False` argument prevents the DataFrame's index from being written to the file, which is often desired when the index is just the default numerical range.



````python
# Save the DataFrame to a CSV file
air_quality_df.to_csv('air_quality_data.csv', index=False)  # index will not be saved
````

### Reading DataFrames from CSV

Just as you can save DataFrames to CSV, you can also load data from existing CSV files into a pandas DataFrame using the `pd.read_csv()` function. This is incredibly useful for importing data that you've collected from various sources or data that you've previously saved.

When you use `pd.read_csv()`, pandas automatically infers the structure of the data, using the first row as column headers by default and determining the data types of each column.

**Footnote** Pandas can also read other file formats generated by excel (.xlsx) using  `pd.read_excel()` function.

````python
# read the csv file we just created:

air_quality_df = pd.read_csv('air_quality_data.csv')
display(air_quality_df)
````

### Example Problem: Analyzing Sales Data

Let's apply what we've learned to a new dataset representing simple sales information.

**Problem:**

You have the following sales data:

| Product | Quantity Sold | Price ($) |
|---|---|---|
| A | 10 | 20.0 |
| B | 5 | 35.0 |
| C | 12 | 15.0 |
| D | 8 | 50.0 |
| E | 15 | 10.0 |

Perform the following steps:

1.  **Create a DataFrame** from this data.
2.  **Calculate the "Total Revenue (\$)"** for each product (Quantity Sold * Price (\$))
4.  **Save the DataFrame** to a new CSV file called `revenue_sales.csv`.
5.  **Read the `revenue_sales.csv`** file back into a new DataFrame and display it.

Try to solve this problem using the pandas methods we've discussed!

````python
import pandas as pd

# Step 1: Create a DataFrame
sales_data = {
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'Quantity Sold': [10, 5, 12, 8, 15],
    'Price ($)': [20.0, 35.0, 15.0, 50.0, 10.0]
}
sales_df = pd.DataFrame(sales_data)

# Step 2: Calculate the "Total Revenue ($)" for each product
sales_df['Total Revenue ($)'] = sales_df['Quantity Sold'] * sales_df['Price ($)']

# Step 3: Save the DataFrame to a new CSV file
sales_df.to_csv('revenue_sales.csv', index=False)

# Step 4: Read the CSV file back into a new DataFrame and display it
revenue_sales_df = pd.read_csv('revenue_sales.csv')
display(revenue_sales_df)
````

### Filtering DataFrames

Filtering is a way to select rows from a DataFrame based on certain conditions. You can filter a DataFrame by using boolean indexing, where you provide a condition that evaluates to `True` or `False` for each row. Only the rows where the condition is `True` will be included in the filtered DataFrame.

For example, we can filter our `air_quality_df` to see only the readings where the temperature in Fahrenheit is above a certain value.

````python
# Filter the DataFrame to show rows where Temperature (°C) is greater than 23
filtered_air_quality_df = air_quality_df[air_quality_df['Temperature (°C)'] > 23]

# Display the filtered DataFrame
display(filtered_air_quality_df)
````
